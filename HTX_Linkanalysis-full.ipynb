# Link Analysis Task I

```python
import numpy as np
import pandas as pd
import networkx as nx
from itertools import combinations
import matplotlib.pyplot as plt

# Data subset (per spec)

df_la = df[df["year"] >= 2017].copy().reset_index(drop=True)

MAX_NODES = len(df_la)
RANDOM_STATE = 42

if len(df_la) > MAX_NODES:
    df_la = df_la.sample(n=MAX_NODES, random_state=RANDOM_STATE).reset_index(drop=True)

print("df_la shape:", df_la.shape)


G = nx.Graph()

# add nodes keep attributes lightweight
for i, r in df_la.iterrows():
    G.add_node(
        i,
        year=int(r["year"]),
        price_psm=float(r["adjusted_price_psm"]),
        town=r["town"],
        flat_type=r["flat_type"],
        block=str(r["block"]),
        street_name=r["street_name"],
        storey_mid=float(r["storey_mid"]) if pd.notna(r["storey_mid"]) else np.nan,
        floor_area_sqm=float(r["floor_area_sqm"]) if pd.notna(r["floor_area_sqm"]) else np.nan,
        remaining_lease_years=float(r["remaining_lease_years"]) if pd.notna(r["remaining_lease_years"]) else np.nan,
        dist_to_city_centre_km=float(r["dist_to_city_centre_km"]) if pd.notna(r["dist_to_city_centre_km"]) else np.nan,
        nearest_mrt_dist_km=float(r["nearest_mrt_dist_km"]) if pd.notna(r["nearest_mrt_dist_km"]) else np.nan,
    )

# 2) Edge rules (fast + explainable)
# Rule 1 core: same block + similar storey
# Rule 2 support: same street + same flat_type + close in year
STOREY_TOL = 5   # floors
YEAR_TOL = 1     # year

# Helper: connect within a small group with constraints
def connect_pairs(indices, rule_name):
    """
    indices: list[int] within a small bucket
    rule_name: edge label
    """
    if len(indices) < 2:
        return 0

    # sort by year to allow early stopping by YEAR_TOL where relevant
    indices = sorted(indices, key=lambda x: df_la.loc[x, "year"])
    added = 0

    # We still do pairwise inside the bucket, but buckets are small.
    for a_pos in range(len(indices)):
        i = indices[a_pos]
        yi = df_la.loc[i, "year"]

        # early stop on year for speed (for rules that use YEAR_TOL)
        for b_pos in range(a_pos + 1, len(indices)):
            j = indices[b_pos]
            yj = df_la.loc[j, "year"]
            if (yj - yi) > YEAR_TOL and rule_name in ("street_type_year",):
                break

            if rule_name == "same_block":
                if abs(df_la.loc[i, "storey_mid"] - df_la.loc[j, "storey_mid"]) <= STOREY_TOL:
                    G.add_edge(i, j, relation=rule_name)
                    added += 1

            elif rule_name == "street_type_year":
                # same street + flat type already guaranteed by groupby
                if abs(yi - yj) <= YEAR_TOL:
                    G.add_edge(i, j, relation=rule_name)
                    added += 1

    return added

#Rule 1: group by block
edges_added_1 = 0
for block, g in df_la.groupby("block", sort=False):
    idxs = g.index.tolist()
    #cap very large blocks (rare) to avoid huge clique
    if len(idxs) > 800:
        idxs = g.sample(n=800, random_state=RANDOM_STATE).index.tolist()
    edges_added_1 += connect_pairs(idxs, "same_block")

print("Edges added (rule 1: same_block):", edges_added_1)

# Rule 2: group by (street_name, flat_type)
edges_added_2 = 0
for (street, ftype), g in df_la.groupby(["street_name", "flat_type"], sort=False):
    idxs = g.index.tolist()
    if len(idxs) > 800:
        idxs = g.sample(n=800, random_state=RANDOM_STATE).index.tolist()
    edges_added_2 += connect_pairs(idxs, "street_type_year")

print("Edges added (rule 2: street+type+year):", edges_added_2)


# High-level graph analysis 
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
avg_degree = np.mean([d for _, d in G.degree()]) if num_nodes > 0 else 0.0
density = nx.density(G) if num_nodes > 1 else 0.0

print("\n--- Graph summary ---")
print(f"Nodes (|V|): {num_nodes}")
print(f"Edges (|E|): {num_edges}")
print(f"Average degree: {avg_degree:.2f}")
print(f"Density: {density:.6f}")

# Connected components
components = list(nx.connected_components(G))
component_sizes = np.array([len(c) for c in components], dtype=int)
largest_cc_size = int(component_sizes.max()) if len(component_sizes) else 0
largest_cc_ratio = (largest_cc_size / num_nodes) if num_nodes else 0.0

print("\n--- Connected components ---")
print(f"Number of components: {len(components)}")
print(f"Largest component size: {largest_cc_size} ({largest_cc_ratio:.1%} of nodes)")

# Degree distribution
degree_series = pd.Series(dict(G.degree()), name="degree")

print("\n--- Degree distribution (summary) ---")
display(degree_series.describe())

# Plots (simple + OA-friendly)
plt.figure()
degree_series.hist(bins=30)
plt.xlabel("Degree")
plt.ylabel("Count")
plt.title("Degree distribution")
plt.show()

plt.figure()
pd.Series(component_sizes).hist(bins=30)
plt.xlabel("Component size")
plt.ylabel("Count")
plt.title("Connected component size distribution")
plt.show()

# quick peek of high-degree nodes 
top_nodes = degree_series.sort_values(ascending=False).head(10).index.tolist()
df_la.loc[top_nodes, ["year", "town", "flat_type", "block", "street_name", "storey_mid",
                      "floor_area_sqm", "adjusted_price_psm"]]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

# largest connected component
largest_cc_nodes = max(nx.connected_components(G), key=len)
G_cc = G.subgraph(largest_cc_nodes).copy()

# keep nodes that actually have edges (avoid isolates)
deg = dict(G_cc.degree())
nodes_with_edges = [n for n, d in deg.items() if d > 0]

# sample nodes from the giant component 
np.random.seed(42)
sample_n = 3000
sample_nodes = np.random.choice(nodes_with_edges, size=min(sample_n, len(nodes_with_edges)), replace=False)

subG = G_cc.subgraph(sample_nodes).copy()

# keep only edges where both endpoints are in the sample (already)
print("subG nodes:", subG.number_of_nodes(), "edges:", subG.number_of_edges())
```

```python
plt.figure(figsize=(12, 9))
pos = nx.spring_layout(subG, seed=42, k=0.35, iterations=200)
pos = nx.spring_layout(subG, seed=42, k=0.15, iterations=400)
# split edges by relation
edges_block = [(u,v) for u,v,d in subG.edges(data=True) if d.get("relation") == "same_block"]
edges_street = [(u,v) for u,v,d in subG.edges(data=True) if d.get("relation") == "street_type_year"]

nx.draw_networkx_nodes(subG, pos, node_size=30, alpha=0.9)

nx.draw_networkx_edges(subG, pos, edgelist=edges_block, width=0.8, alpha=0.35)
nx.draw_networkx_edges(subG, pos, edgelist=edges_street, width=0.5, alpha=0.20)

plt.title("Listing Similarity Graph (sampled from largest component)\nEdges: same_block vs street_type_year")
plt.axis("off")
plt.show()
```

# Link Analysis Task I Intepretation

We construct a graph that links property transactions which are comparable, similar to how buyers compare flats when forming price expectations.

Nodes. Each node represents a resale transaction record (Jan 2017 onwards). Node attributes include town, flat type, block/street, storey level, floor area, remaining lease, distance to MRT / city centre, and the target variable adjusted_price_psm (used for analysis, not for edge creation).

Edges (rule-based, interpretable). We connect two nodes if they are likely comparable based on:

* Same block + similar storey level (|storey_mid_i - storey_mid_j| â‰¤ 5)
Rationale: flats within the same block share highly similar micro-location and amenities; restricting storey difference prevents linking low- and high-floor units that behave differently.

* Same street + same flat type + close in time (|year_i - year_j| â‰¤ 1)
Rationale: similar unit type within the same street and close time period reflects the same micro-market and buyer comparisons.

This rule-based approach is intentionally chosen over embedding/KNN graphs to remain transparent, reproducible, and CPU-friendly for teammates without GPU access.

# Link Analysis Task II

```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

def get_node_price(G, n):
    d = G.nodes[n]
    if "price_psm" in d:
        return d["price_psm"]
    return np.nan
```

## algorithm 1ï¼šConnected Components + Component-level price summary

### Connected components

```python

components = list(nx.connected_components(G))
comp_sizes = np.array([len(c) for c in components])

print(f"Number of components: {len(components)}")
print(f"Largest component size: {comp_sizes.max()} ({comp_sizes.max()/G.number_of_nodes():.1%} of nodes)")

# Component size distribution
plt.figure(figsize=(8,4))
plt.hist(comp_sizes, bins=30)
plt.title("Connected component size distribution")
plt.xlabel("Component size")
plt.ylabel("Count")
plt.show()

# Component-level price statistics 
comp_rows = []
for k, comp in enumerate(components):
    nodes = list(comp)
    prices = np.array([get_node_price(G, n) for n in nodes], dtype=float)
    prices = prices[~np.isnan(prices)]
    comp_rows.append({
        "component_id": k,
        "size": len(nodes),
        "mean_price_psm": float(np.mean(prices)) if len(prices) else np.nan,
        "std_price_psm": float(np.std(prices)) if len(prices) else np.nan,
    })

comp_df = pd.DataFrame(comp_rows).sort_values(["size"], ascending=False)

#comp_df = comp_df.sort_values(by=['mean_price_psm'], ascending=False)

comp_df
```

### Component size vs mean price

```python
#Component size vs mean price
plt.figure(figsize=(8,4))
plt.scatter(comp_df["size"], comp_df["mean_price_psm"], alpha=0.6)
plt.title("Component size vs mean price_psm")
plt.xlabel("Component size")
plt.ylabel("Mean adjusted_price_psm")
plt.show()
```

```python
from networkx.algorithms.community import greedy_modularity_communities

# subG should be a sampled subgraph ( 3000 nodes)
comms = list(greedy_modularity_communities(subG))
print("num communities in subG:", len(comms))

node_to_comm = {}
for i, c in enumerate(comms):
    for n in c:
        node_to_comm[n] = i

# Plot community-colored
colors = [node_to_comm.get(n, 0) for n in subG.nodes()]

plt.figure(figsize=(12,9))
pos = nx.spring_layout(subG, seed=42, k=0.25, iterations=200)
nx.draw_networkx_nodes(subG, pos, node_size=25, node_color=colors, cmap="tab20", alpha=0.9)
nx.draw_networkx_edges(subG, pos, width=0.5, alpha=0.2)
plt.title("Community structure on sampled subgraph (greedy modularity)")
plt.axis("off")
plt.show()
```
